```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,14,0.1)
y = np.cos(x) -np.sin(x)
y_1 = np.cos(x) 
y_2 = np.cos(x) +np.sin(x)
```

```python
plt.plot(x,y, label = "v0<0" )
plt.plot(x,y_1, label = "vo=0")
plt.plot(x,y_2, label="vo>0")
plt.legend()
```
