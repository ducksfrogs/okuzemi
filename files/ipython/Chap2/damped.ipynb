```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10,0.1)
```

```python
def real_root(x):
    return -0.02*np.exp(-9*x) + 0.18*np.exp(-x)
```

```python
y = real_root(x)
```

```python
plt.plot(x,y)
```

```python
def double_root(x):
    return (0.16 + 0.48*x)*np.exp(-3*x)
```

```python
y2 = double_root(x)
```

```python
plt.plot(x,y)
plt.plot(x,y2)
```
