```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(.1,100,.1)
```

```python
y = x**(1/3)*(1-1/3*np.log(x))
```

```python
plt.plot(x,y, label='0<x<100')
plt.legend()
plt.grid()
plt.savefig("0_100.png")
```

```python
y = x**(1/3)*(1+1/3*np.log(x))
```

```python
x = np.arange(.1,7,.1)
y = x**(1/3)*(1+1/3*np.log(x))
```

```python
plt.plot(x,y, label="0<x<7")
plt.legend()
plt.grid()
plt.savefig("0_7.png")
```
