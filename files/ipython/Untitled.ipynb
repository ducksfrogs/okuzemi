```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.plot([1,2,3,4])
plt.ylabel('some number')
plt.show()
```

```python
x = np.arange(0,2,0.1)
y = np.exp(x)+3*np.exp(-2*x)
```

```python
plt.plot(x,y)
plt.savefig("real.png")
```

```python
x= np.arange(0,14,0.1)
y = np.exp(-0.2*x)*np.sin(3*x)
```

```python
plt.plot(x,y)
plt.savefig("imag.png")
```

```python
y=(3-2*x)*np.exp(-0.5*x)
```

```python
plt.plot(x,y)
plt.grid()
plt.savefig("double.png")
```
