```python
x = np.arange(0,5,0.1)
```

```python
y =(1-1*x)*np.exp(-x)
y_2 =(1-0.5*x)*np.exp(-x)
y_3=(1-0.333*x)*np.exp(-x)
y_4 =(1-0.25*x)*np.exp(-x)
```

```python
plt.plot(x,y, label="t=-2")
plt.plot(x,y_2, label="t=-1.5")
plt.plot(x,y_3, label="t=-1.3")
plt.plot(x,y_4, label="t=-1.25")
plt.legend()
plt.grid()
```
